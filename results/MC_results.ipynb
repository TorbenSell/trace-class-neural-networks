```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
KL_steps=np.load('KL_steps.npy',allow_pickle=True)
BNN_steps=np.load('BNN_steps.npy',allow_pickle=True)
tcNN_steps=np.load('tcNN_steps.npy',allow_pickle=True)

steps = []
steps.append(KL_steps[0]) # optimal policy
steps.append(KL_steps[1]) # KL pCN
steps.append(KL_steps[2]) # KL pCNL
steps.append(BNN_steps[1]) # NN pCN
steps.append(BNN_steps[2]) # NN pCNL
steps.append(tcNN_steps[1]) # NN pCN
steps.append(tcNN_steps[2]) # NN pCNL

KL_fails=np.load('KL_failures.npy')
BNN_fails=np.load('BNN_failures.npy')
tcNN_fails=np.load('tcNN_failures.npy')
fails=np.zeros(7)
fails[0] = KL_fails[0] # optimal policy
fails[1] = KL_fails[1] # KL pCN
fails[2] = KL_fails[2] # KL pCNL
fails[3] = BNN_fails[1] # NN pCN
fails[4] = BNN_fails[2] # NN pCNL
fails[5] = tcNN_fails[1] # NN pCN
fails[6] = tcNN_fails[2] # NN pCNL
print(fails)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Steps until success for different policies')
ax.boxplot(steps)
ax.set_xticklabels(['optimal','KL, pCN','KL, pCNL','BNN, pCN','BNN, pCNL','tcNN, pCN','tcNN, pCNL'])
plt.show()

fig.savefig('MC_results.pdf', dpi=300, bbox_inches='tight')
plt.close(fig) 
```
