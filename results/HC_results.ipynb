```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
KL_dist=np.load('KL_dist.npy',allow_pickle=True)
BNN_dist=np.load('BNN_dist.npy',allow_pickle=True)
tcNN_dist=np.load('tcNN_dist.npy',allow_pickle=True)

dist = []
dist.append(KL_dist[0]) # optimal policy
dist.append(KL_dist[1]) # KL pCN
dist.append(KL_dist[2]) # KL pCNL
dist.append(BNN_dist[0]) # NN pCN
dist.append(BNN_dist[1]) # NN pCNL
dist.append(tcNN_dist[0]) # NN pCN
dist.append(tcNN_dist[1]) # NN pCNL
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_title('Distances covered for different policies')

ax.plot(dist[0], len(dist[0]) * [0], "x")
ax.plot(dist[1], len(dist[1]) * [-1], "x")
ax.plot(dist[2], len(dist[2]) * [-2], "x")
ax.plot(dist[3], len(dist[3]) * [-3], "x")
ax.plot(dist[4], len(dist[4]) * [-4], "x")
ax.plot(dist[5], len(dist[5]) * [-5], "x")
ax.plot(dist[6], len(dist[6]) * [-6], "x")

ax.set_yticks([0,-1,-2,-3,-4,-5,-6])
ax.set_yticklabels(['optimal','KL, pCN','KL, pCNL','BNN, pCN','BNN, pCNL','tcNN, pCN','tcNN, pCNL'])
plt.show()

fig.savefig('HC_results.pdf', dpi=300, bbox_inches='tight')
plt.close(fig) 
```
